### Sorting Data

```python
import pandas as pd
```

```python
people = {
    "first" : ["Vijay", "Abhishek", "Ravi", "Anand", "John", "Shivam", "Neeraj", "Shivansh", "Akash", "Saurabh"],
    "last" : ["Yadav", "Yadav", "Singh", "Pal", "Doe", "Bind", "Yadav", "Singh", "Singh", "Maurya"],
    "email" : ["vijay@gmail.com", "abhi@gmail.com", "ravi@gmail.com", "anand@gmail.com", "john@gmail.com", "shivam@gmail.com", "neeraj@gmail.com", "shiv@gmail.com", "akash@gmail.com", "saurabh@gmail.com"],
    "qualification" : ["BSc", "BCA", "B.A.", "B.Tech", "M.Tech", "M.A.", "BSc", "BCA", "LLB", "BCA"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='first')
```

```python
df.sort_values('first', ascending=False)
```

```python
df.sort_values(['first', 'last'])
```

```python
# First name in descending order and last name in ascending order 
df.sort_values(by=['first', 'last'], ascending=[False, True])
```

```python
# Sorting by index 
df.sort_index()
```

```python
df['last'].sort_values()
```
