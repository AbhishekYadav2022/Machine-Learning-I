### Add/Remove Rows and Columns from DataFrames

```python
import pandas as pd
```

```python
people = {
    "first" : ["Vijay", "Abhishek", "Ravi", "Anand", "John", "Shivam", "Neeraj", "Shivansh", "Akash", "Saurabh"],
    "last" : ["Yadav", "Yadav", "Singh", "Pal", "Doe", "Bind", "Yadav", "Singh", "Singh", "Maurya"],
    "email" : ["vijay@gmail.com", "abhi@gmail.com", "ravi@gmail.com", "anand@gmail.com", "john@gmail.com", "shivam@gmail.com", "neeraj@gmail.com", "shiv@gmail.com", "akash@gmail.com", "saurabh@gmail.com"],
    "qualification" : ["BSc", "BCA", "B.A.", "B.Tech", "M.Tech", "M.A.", "BSc", "BCA", "LLB", "BCA"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['first'] + ' ' + df['last']
```

```python
df['full_name'] = df['first'] + ' ' + df['last']
```

```python
df
```

```python
df.drop(columns=['first', 'last'], inplace=True)
```

```python
df
```

```python
# Reverse
df[['first', 'last']] = df['full_name'].str.split(' ', expand=True)
```

```python
df
```

```python
df.append({'first': 'Tony'}, ignore_index=True)
```

```python
people = {
    "first" : ["Tony", "Steve"],
    "last" : ["Stark", "Rogers"],
    "email" : ["stark@gmail.com", "steve@gmail.com"],
    "qualification" : ["B.Tech", "MBA"]
}
df2 = pd.DataFrame(people)
```
