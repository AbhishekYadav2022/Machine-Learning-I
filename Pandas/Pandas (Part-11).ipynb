#### Exporting Data To SQL Database

```python
from sqlalchemy import create_engine, text
import psycopg2
```

```python
engine = create_engine('postgresql://postgres:123@localhost:5432/online_shopping_app')
```

```python
# india_df.to_sql('sample_table', engine)
```

#### Importing Data From SQl Database

```python
with engine.begin() as conn:
    sql_df = pd.read_sql('users', conn) 
```

```python
sql_df
```

```python
with engine.begin() as conn:
    query = text("""SELECT * FROM users""")
    sql_df = pd.read_sql_query(query, conn)
# OR
# sql_df = pd.DataFrame(engine.connect().execute(text('SELECT * FROM users')))
```
