### Updating Rows and Columns
Modifying Data Within DataFrames

```python
import pandas as pd
```

```python
people = {
    "first" : ["Vijay", "Abhishek", "Ravi", "Anand", "John", "Shivam", "Neeraj", "Shivansh", "Akash", "Saurabh"],
    "last" : ["Yadav", "Yadav", "Singh", "Pal", "Doe", "Bind", "Yadav", "Singh", "Singh", "Maurya"],
    "email" : ["vijay@gmail.com", "abhi@gmail.com", "ravi@gmail.com", "anand@gmail.com", "john@gmail.com", "shivam@gmail.com", "neeraj@gmail.com", "shiv@gmail.com", "akash@gmail.com", "saurabh@gmail.com"],
    "qualification" : ["BSc", "BCA", "B.A.", "B.Tech", "M.Tech", "M.A.", "BSc", "BCA", "LLB", "BCA"]
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.columns
```

```python
df.columns = ['first_name', 'last_name', 'email', 'qualification']
```

```python
df
```

```python
# df.columns = [x.upper() for x in df.columns]
df.columns = [x.lower() for x in df.columns]
```

```python
df
```

```python
# df.columns = df.columns.str.replace('_', ' ')
```

```python
df
```

```python
df.rename(columns={'first_name':'first', 'last_name':'last'}, inplace=True)
```

```python
df
```

```python
# Modifying data of one row
df.loc[1] = ['Abhishek', 'Yadav', 'abhi@gmail.com', 'BCA']
df
```

```python
df.loc[1, ['email']] = ['abhishek@gmail.com']
```

```python
df
```

```python
# df.loc[1, 'first'] = 'Abhi'
# Or
df.at[1, 'first'] = 'Abhi'
```

```python
df
```

Modifying Column Data

```python
# df['first'] = df['first'].str.upper()
df
```

```python
df['first'].apply(len)
```

```python
def update_email(email):
    return email.upper()
```

```python
df['email'].apply(update_email)
```

```python
# Using Lambda Function
df['email'] = df['email'].apply(lambda x: x.lower())
```

```python
df
```

```python
df.apply(len)
```

```python
df.apply(len, axis=1)
```

```python
len(df['email'])
```

```python
df.apply(pd.Series.min)
```

```python
# Applying a function to every element of dataframe 
df.applymap(len)
```

```python
df.applymap(str.lower)
```
